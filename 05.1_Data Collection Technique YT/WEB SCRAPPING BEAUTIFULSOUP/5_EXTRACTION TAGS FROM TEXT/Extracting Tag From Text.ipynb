```python
import requests
from bs4 import BeautifulSoup
```

```python
web = requests.get("https://tutorialsfreak.com/")
web
```

```python
soup = BeautifulSoup(web.content, "html.parser")
```

```python
lines = soup.find_all("p")
```

```python
for l in lines:
    print(l.text)
```

```python
s = soup.find("div", class_ = "why-choose-card card-shadow card")
```

```python
for l1 in lines_1:
    print(l1.text)
```
