```python
import pandas as pd

#makeing sereis and giving index...
s2 = pd.Series([90,94,56,78,90,32],index = ["Anas","Juniad","Kirti","Shubham","Arif","Ram"])

#Printing the series and data type...
print(s2)
print("Type:",type(s2))
```

DataFrame — 2D Labeled Table

```python
#Makeing Dataframe..
data = {
    "Name":["Arif","Anas","Junaid"],
    "Age":[23,22,25],
    "City":["Bokaro","Garia","Dhanbad"]
}

#Printing the dataframe..
df = pd.DataFrame(data)

df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]

print(df)
print("\nIndexs:",df.index)
print("Column:",df.columns)
```
