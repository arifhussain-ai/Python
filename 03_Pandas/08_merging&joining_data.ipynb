```python
import pandas as pd
```

```python
Employee = pd.DataFrame({
    "Emply_Id": [1,2,3],
    "Name": ["Alice","Bob","Charlie"],
    "Dept_Id": [10,20,30] 
})
```

```python
Employee
```

```python
Department = pd.DataFrame({
    "Dept_Id": [10,20,40],
    "Dept_Name": ["HR","Engneering","Marketing"] 
})
```

```python
Department
```

```python
pd.merge(Employee, Department, on = "Dept_Id") #Inner join
```

```python
pd.merge(Employee, Department, on = "Dept_Id", how= "left") #left join
```

```python
pd.merge(Employee, Department, on = "Dept_Id", how= "right") #Right Join
```

```python
pd.merge(Employee, Department, on = "Dept_Id", how="outer") #Outer Join
```
