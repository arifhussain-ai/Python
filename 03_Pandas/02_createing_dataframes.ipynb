```python
import pandas as pd

#Createing data from python list.....
data = [["Arif",90],
        ["Jacob",87],
        ["Jack",48],
        ["Jerry",56]]

#Printing the dataframe...
df = pd.DataFrame(data,columns=["Name","Marks"])
print(df)
```

```python
import pandas as pd

#Createing data from Dictionary list.....
data = {
    "Name":["Arif","Anas","Junaid"],
    "Course":["BAC","BTECH","PHY EDU"],
    "Year":[2023,2024,2022]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Createing dataframe with the help of numpy array....
import numpy as np

arr = np.array([[1,90],
                [2,89],
                [3,78]])

data = pd.DataFrame(arr,columns=["Roll","Marks"])
data.index = ["a","b","c"]

print(data)
```

```python
data = [["Alice",65],
        ["Bob",87],
        ["Charlie",45]]

df = pd.DataFrame(data,columns=["Name","Age"])
print(df)
```

```python
data = {
    "Name":["Alice","Bob","Charlie"],
    "Age":[65,87,45]
}

df = pd.DataFrame(data)
print(df)
```
