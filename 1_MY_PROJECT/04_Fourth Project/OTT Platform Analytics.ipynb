```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

```python
type_count = df["type"].value_counts()
plt.figure(figsize= (9,6))

sns.set(style = "darkgrid")

sns.barplot(x = type_count.index, y = type_count.values,palette= "Set2")
plt.title("Movies vs TV Shows Count", fontsize = 20, fontweight = "bold")
plt.xlabel("Content Type", fontsize = 11)
plt.ylabel("Count", fontsize = 11)

plt.show()
```

```python
country_count = df["country"].value_counts().head(10)
plt.figure(figsize= (10,6))

sns.set(style = "dark")

sns.barplot(x = country_count.index, y = country_count.values, palette= "pastel")
plt.title("Country Wise Content Distribution", fontsize = 26, fontweight = "bold")
plt.xlabel("Country", fontsize = 11)
plt.ylabel("Count", fontsize = 11)
plt.xticks(rotation = 45)

plt.show()
```

```python
genres_count = df["listed_in"].str.split(", ").explode().value_counts().head(10)
plt.figure(figsize= (10,6))

sns.set(style = "darkgrid")

sns.barplot(x = genres_count.index, y = genres_count.values, palette= "coolwarm")
plt.title("Genre Popularity Analysis", fontsize = 25, fontweight = "bold")
plt.xlabel("Genres", fontsize = 13)
plt.ylabel("Count", fontsize = 13)
plt.xticks(rotation = 45)

plt.show()
```

```python
movie_df = df[df["type"] == "Movie"].copy()

movie_df["duration"] = movie_df["duration"].str.replace("min","")
movie_df = movie_df.dropna(subset = ["duration"])
movie_df["duration_int"] = movie_df["duration"].astype(int)

plt.figure(figsize= (10,6))

sns.set(style = "darkgrid")

sns.histplot(movie_df["duration_int"], kde= True, bins = 20, color = "g")
plt.title("Duration Distribution", fontsize = 30, fontweight = "bold")
plt.xlabel("Duration Time", fontsize = 13)
plt.ylabel("Count", fontsize = 13)

plt.show()
```

```python
year_count = df["release_year"].value_counts().sort_index()
plt.figure(figsize= (12,6))

sns.set(style = "darkgrid")

sns.lineplot(x = year_count.index, y = year_count.values,color = "g")
plt.title("Year Wise Content Growth", fontsize = 30, fontweight = "bold")
plt.xlabel("Release Year", fontsize = 13)
plt.ylabel("Count", fontsize = 13)

plt.show()
```

```python
top_directors = df["director"].value_counts().head(5)
plt.figure(figsize= (12,6))

sns.set(style = "darkgrid")

sns.barplot(x = top_directors.index, y = top_directors.values, palette= "pastel")
plt.title("Top 5 Directors Analysis", fontsize = 30, fontweight = "bold")
plt.xlabel("Dierctors", fontsize = 12)

plt.show()
```

```python
rating_count = df["rating"].value_counts()
plt.figure(figsize= (12,6))

sns.set(style = "darkgrid")

sns.barplot(x = rating_count.index, y = rating_count.values, palette = "Set2")
plt.title("Most Common Rating Category", fontsize = 25, fontweight = "bold")
plt.xlabel("Rating", fontsize = 12)

plt.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors = "coerce")
df["year_added"] = pd.to_datetime(df["date_added"]).dt.year

growth = df["year_added"].value_counts().sort_index()
plt.figure(figsize= (12,6))

sns.set(style = "darkgrid")
sns.lineplot(x = growth.index, y = growth.values, color = "g", marker = "o")
plt.title("Content Addition Trend", fontsize = 30, fontweight = "bold")
plt.xlabel("Years", fontsize = 12)

plt.show()
```

```python
plt.figure(figsize= (10,6))

sns.set(style = "darkgrid")
sns.countplot(data = df, x = "type", palette="Set2")
plt.title("Movie vs TV Show Duration Comparison", fontsize = 22, fontweight = "bold")
plt.xlabel("Type Of Contents", fontsize = 12)

plt.show()
```

```python
plt.figure(figsize= (10,6))
pivt = pd.pivot_table(df, index = "country", columns= "type", aggfunc = "size")

pivt["Total"] = pivt["Movie"] + pivt["TV Show"]
pivt = pivt.sort_values(by = "Total", ascending= False).head(10)
sns.heatmap(pivt, annot = True, cmap = "coolwarm")

plt.title("Country vs Content Type Heatmap", fontsize = 21, fontweight = "bold")

plt.show()
```
