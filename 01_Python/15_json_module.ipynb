```python
import json
```

```python
obj = {"Name": "Arif", "Company": "Microsoft", "Salary": 450000}
s = json.dumps(obj)
print(s)
print("Type:",type(s))
```

```python
file = open("15.1_data.json","w")
json.dump(obj,file)
file.close()
```

```python
formatted_json = json.dumps(obj,indent = 2)
print(formatted_json)
```

```python
string_data = '{"Name": "Arif", "Company": "Microsoft", "Salary": 450000}'
python_obj = json.loads(string_data)

print(python_obj)
print(type(python_obj))
```

```python
with open ("15.1_data.json" , "r") as file:
    python_obj = json.load(file)
print(python_obj)
```
