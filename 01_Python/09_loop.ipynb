```python
fruits = ['Apple',"Banana","Mango"]
for fruit in fruits:
    if(fruit == "Banana"):
        break #Break the loop instantly...
    print(fruit)
```

```python
fruits = ['Apple',"Banana","Mango"]
for fruit in fruits:
    if(fruit == "Banana"):
        continue #Skip the given item and print the next item...
    print(fruit)
```
