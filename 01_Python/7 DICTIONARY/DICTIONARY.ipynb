```python
marks = {"ARIF":99,"RAMESH":89,"SURESH":90}
max_marks = max(marks.values())

for name,score in marks.items():
    if score == max_marks:
        print("TOPPER:",name,"=",score)
```

```python
s = "banana"
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

```python
lst = [1,1,2,2,2,3]
freq = {}

for i in lst:
    freq[i] = freq.get(i,0) + 1

group = {}
for key,value in freq.items():
    if value not in group:
        group[value] = []
    group[value].append(key)

print(group)
```

```python
d = {"a": 1, "b": 2, "c": 1, "d": 2}
result = {}

for k,v in d.items():
    result.setdefault(v, []).append(k)

print(result)
```

```python
d = {"a": 1, "b": {"c": 2, "d": 3}}
flat = {}

for k,v in d.items():
    if isinstance(v, dict):
        for k2, v2 in v.items():
            flat[k2] = v2
    else:
        flat[k] = v
        
print(flat)
```
