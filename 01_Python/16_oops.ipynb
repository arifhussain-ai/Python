```python
class Employee:
    company = "HP"
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        
    def printdetails(self):
        print(f"The company of {self.name} name is {self.company} with salary ${self.salary}")

    @staticmethod
    def printtime():#staticmethod will that it run this function without self because this func nothing to do with self
        print("The time is now...")

    @classmethod
    def classdetails(cls):
        print(f"The comany name is {cls.company}")
        
e = Employee("Arif",450000)
e.printdetails()
Employee.printtime()
Employee.classdetails()
```

```python
class Car:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def printinfo(self):
        print(f"The Car name is: {self.name} and The Brand name is: {self.brand}")

c = Car("Scorpio", "Mahindra")
c.printinfo()
```

```python
class BankAmount:
    def __init__(self,balance):
        self.__balance = balance  #Private attribute

    def deposite(self,amount):  #Setter.....
        if amount > 0:
            self.__balance += amount

    def get_balance(self):     #Getter.....
        return self.__balance

account = BankAmount(100000)
account.deposite(5000)
print(f"The Bank Balance :",account.get_balance())
```

```python
class A:
    def method_a(self):
        return "METHOD A...!"

class B:
    def mehtod_b(self):
        return "METHOD B...!"

class C(A,B):     #Multiple inheritance
    pass

obj = C()
print(obj.method_a())
print(obj.mehtod_b())
```

```python
#Polymorphism method....
class Bird:
    def fly(self):
        return "Brid can Fly"
                                  #In this method it call indivisual class
class Penguin:
    def fly(self):
        return "Penguin can Fly"

bird = Bird()
penguin = Penguin()
print(bird.fly())
print(penguin.fly())
```

```python
#A.bstraction is used to define a method without implementing it in the base class.
#It is achieved using abstract base classes ( ABC module).
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass #Nothing do in this class it will do in child class...

class Square(Shape):
    def __init__(self,side):
        self.side = side

    def area(self):
        return self.side * self.side #implement in child class from parent class..

sqr = Square(4)
print("Side Of Square: ",sqr.area())
```

```python
#Magic Method
class Books:
    def __init__(self,tittle,pages):
        self.tittle = tittle
        self.pages = pages

    def __str__(self):
        return f"Book: {self.tittle}"

    def __len__(self):
        return self.pages


books = Books("Pyhton Basic",350)
print(str(books))
print(f"Pages:",len(books))
```

```python
class Example:
    cls_var = "Class Method"

    @classmethod
    def class_var(cls):
        return cls.cls_var

    def instant_var(self):
        return "Instant Variable"

    @staticmethod
    def static_var():
        return "Static Variable"

obj = Example()
print(Example.class_var())
print(obj.instant_var())
print(Example.static_var())
```
