```python
square = [x**2 for x in range(5)]
print(square)
```

```python
Evens = [x for x in range(10) if x%2 == 0]
print(Evens)
```

```python
numbers = ["Even" if x%2 == 0 else "Odd" for x in range(11)]
print(numbers)
```

```python
pairs = [(x,y) for x in range(2) for y in range(2)]
print(pairs)
```

```python
words = ["hello","python","world"]
upper_words = [word.upper() for word in words]
print(upper_words)
```

```python
matrix = [[1,2],[3,4],[5,6]]
fallented = [num for row in matrix for num in row]
print(fallented)
```

```python
unique_num = {x for x in [1,2,2,3,3,4]}
print(unique_num)
```

```python
square_dict = {x : x**2 for x in range(1,5)}
print(square_dict)
```

```python
import time

start = time.time()
square_loop = []
for x in range(10**2):
    square_loop.append(x**2)
print("Loop Time:",time.time() - start)
```

```python
start = time.time()
square_lst = [x**2 for x in range(10**6)]
print("List Comprehension Time:",time.time() - start)           
```
