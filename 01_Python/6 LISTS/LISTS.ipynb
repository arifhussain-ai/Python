```python
lst = [10,5,6,9,8,10,7]
new_lst = []

for i in lst:
    if i not in new_lst:
        new_lst.append(i)
        
print(new_lst)
```

```python
lst = [1,1,2,2,2,3]
freq = {}

for i in lst:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print(freq)
```

```python
lst = [1,2,3,4,5]
rev1 = []

for i in lst:
    rev1 = [i] + rev1
print("ANSWER 1:",rev1)

#ALTERNATIVE AND EASYWAY....
rev2 = lst[::-1]
print("ANSWER 2:",rev2)
```

```python
lst = [2, 4, 3, 5, 7, 8]
target = 8

for i in range(len(lst)):
    for j in range(i+1, len(lst)):
        if lst[i] + lst[j] == target:
            print("TARGET NUIMBER:",lst[i], lst[j])
```

```python
lst = [1, 2, 3, 4, 5, 7]   # missing 6
n = 7

total = n * (n + 1) // 2
missing = total - sum(lst)

print("MISSING:",missing)
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]

seen = set()
duplicate = set()

for i in lst:
    if i in seen:
        duplicate.add(i)
    else:
        seen.add(i)

print("DUPLICATE NUMBERS:",duplicate)
```
