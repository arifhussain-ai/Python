```python
import numpy as np
import time
```

Built-in NumPy methods are 100-1000 times faster than Python loop Implementation

```python
#test and compare performance of numpy vs plain python loop - dot product
n = 10000000
```

```python
#vector
n = 10000000
a = [1] * n
b = [2] * n

start = time.time()
c = [0] * n

for i in range(n):
    c[i] = a[i] + b[i]

end = time.time()

print("Time taken from Vector:",end-start)
```

```python
#Numpy
start = time.time()

a_np = np.arange(n) * 1
b_np = np.arange(n) * 2

c_np = a_np + b_np

end = time.time()

print("Time taken from Numpy:",end-start)
```

```python
# create 2 matrices using numpy array
```

```python
A = np.array([[1,2,3],[0,1,0],[0,0,1]]) #matrix = A
B = np.array([[0,1,0],[1,0,0],[1,2,1]])  #matrix = B
```

```python
print("Shape of A:",A.shape)
print("Shape of B:",B.shape)
```

```python
A
```

```python
B
```

```python
# find sum of the matrices
```

```python
print(A + B)
```

```python
# find difference of the matrices
```

```python
print(A - B)
```

```python
# finding product of 2 matrices
```

```python
print(A @ B)
print(np.matmul(A,B))
```

```python
# finding transpose of a matrix
```

```python
print(A.T)
print(np.transpose(A))
```

```python
# finding determinant of a matrix
```

```python
print(np.linalg.det(A))
print(np.linalg.det(B))
```

```python
# finding inverse of a matrix
```

```python
print(np.linalg.inv(A))
print(np.linalg.inv(B))
```

```python
# finding trace of a matrix
```

```python
print(np.linalg.trace(A))
print(np.linalg.trace(B))
```

```python
# finding rank of a matrix
```

```python
print(np.linalg.matrix_rank(A))
print(np.linalg.matrix_rank(B))
```

```python
# finding eigen value & Eigen vector of a matrix
```

```python
print(np.linalg.eig(A))
print(np.linalg.eig(B))
```
