```python
import numpy as np

arr = np.array([1,2,3,4,5,6,7])
print("Data_Type:",arr.dtype)
```

```python
#Changing the data type...
float_arr = np.array([1.2,3.5,2.6])
print("Arrays:",float_arr)
print("Data_type:",float_arr.dtype)

#Converting the data_type....
int_arr = float_arr.astype("int64")
print("\nArrays:",int_arr)
print("Data_type:",int_arr.dtype)

#Makeing new array using the dtype...
arr2 = np.array([1,2,3,4,5,6,7], dtype = "float32")
print("\nArray:",arr2)
```

```python
# Downcasting to Save Memory
arr_large = np.array([100000,200000,300000], dtype="int64")

arr_smaller = arr_large.astype("int32")
print("Data:",arr_smaller)
print("Data Type:",arr_smaller.dtype)
```

```python
#Example: Memory Usage
arr_int64 = np.array([1,2,3], dtype = "int64")
arr_int32 = np.array([1,2,3],dtype = "int32")

#Storage taken....
print("Sizen taken:",arr_int64.nbytes)
print("Sizen taken:",arr_int32.nbytes)
```

```python
#String Array
arr_str = np.array(["Apple","Banana","Graphes","Cherry"], dtype= "str")

print("Arrays:",arr_str)
print("Data tpye:",arr_str.dtype)
```

```python
#Complex Numbers

complex_arr = np.array([1 + 2j, 3 + 4j, 5 + 6j, 7 + 8j], dtype = "complex128")

print("Complex_Array:",complex_arr)
print("Data type:",complex_arr.dtype)
```

```python
# Object Data Type
arr_obj = np.array([{"a":1}, [1,2,3], "hello"], dtype = "object")

print("Object Array:", arr_obj)
print("Data type:",arr_obj.dtype)
```
