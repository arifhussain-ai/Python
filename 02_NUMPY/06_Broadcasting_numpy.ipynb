```python
import numpy as np
import time

# Looping Over Arrays in Python
arr = np.array([1,2,3,4,5,6])
result = []

#using loop square the array..
start = time.time()
for num in arr:
    result.append(int(num ** 2))
end = time.time()

#Printing the result....
print("Result:",result)
print("Time taken:",end-start)
```

```python
#Vectorized Operation
arr = np.array([1,2,3,4,5,6])

start = time.time()
result = arr**2
end = time.time()

print("Result:",result)
print("Time taken:",end-start)
```

```python
#Broadcasting with Scalar
arr = np.array([1,2,3,4,5,6])
result = arr + 10
print("Array:",arr)

print("Output:",result)
```

```python
# Broadcasting with Two Arrays
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

result = arr1 + arr2
print("Result:",result)
```

```python
#Broadcasting a 2D Array and a 1D Array

arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([1,2,3])

result = arr1 + arr2
print("Result:",result)
```

```python
# Simulating a dataset (5 samples, 3 features)
data = np.array([[10, 20, 30],
[15, 25, 35],
[20, 30, 40],
[25, 35, 45],
[30, 40, 50]])

# Calculating mean and standard deviation for each feature (column)
mean = data.mean(axis = 0)
std = data.std(axis = 0)

# Normalizing the data using broadcasting
normalized_data = (data - mean) / std

print("Normalized data:",normalized_data)
```
