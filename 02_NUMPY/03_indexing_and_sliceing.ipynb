#Fancy Indexing & Boolean Masking

```python
arr = np.array([10,25,48,96,74,10])
idx = [0,4,5]
print(arr[idx])
```

```python
arr = np.array([10,25,48,96,74,10])
msk = arr > 50
print(arr[msk])
```
