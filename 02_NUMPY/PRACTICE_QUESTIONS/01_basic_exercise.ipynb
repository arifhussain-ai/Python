```python
import numpy as np
```

```python
#create an array

arr = np.array([1,2,3,4,5,6])

#Printing arrays...
print("Arrays:",arr)
```

```python
#Create a 2D Array

arr2D = np.array([[1,2,3],
                  [4,5,6]])

#printing 2D array...
print("2D Arrays:",arr2D)
```

```python
#Array of Zeros
zeros = np.zeros(5)

#print array..
print("Zero array:",zeros)
```

```python
#Create a 3×3 array filled with ones.
ones = np.ones((3,3))

#printing arrays
print("Ones Arrays:",ones)
```

```python
#Create an array with numbers from 0 to 9

arr = np.arange(0,10)

#print arrays...
print("arrays:",arr)
```

```python
#Find the shape of the array np.array([[1,2,3],[4,5,6]])
arr = np.array([[1,2,3],[4,5,6]])

#printing arrays shape...
print("Shapes:",arr.shape)

#Check the data type of an array
print("Data type:",arr.dtype)
```

```python
#Reshape an array of numbers from 1 to 6 into a 2×3 array

arr = np.arange(1,7)

#Reshapeing array..
reshape = arr.reshape(2,3)

#printing arrays..
print("Reshape Arrays:",reshape)
```

```python
#Access the element 5 from the array [[1,2,3],[4,5,6]].
arr = np.array([[1,2,3],[4,5,6]])

#Access the element 5
print("Access the element 5:",arr[1,1])
```

```python
#Access the element 5 from the array [[1,2,3],[4,5,6]]
arr = np.array([[1,2,3],[4,5,6]])

#printing the element 5
print("Element 5:",int(arr[1, 1]))
```

```python
#Get the first three elements from array [10, 20, 30, 40, 50].
arr =  np.array([10, 20, 30, 40, 50])

idx = [0,1,2]

print(arr[idx])
```

```python
#Add two arrays [1,2,3] and [4,5,6]

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

arr = arr1 + arr2

print(arr)
```

```python
#Multiply all elements of an array by 2.
arr = np.array([1,2,3])

arr1 = arr*2

print(arr1)
```

```python
#Find the mean of array [1,2,3,4,5]
array_1 = np.array([1,2,3,4,5])

array_mean = np.mean(array_1)
print(int(array_mean))
```

```python
#Find the maximum and minimum values of an array.
print(int(np.min(array_1)))
print(int(np.max(array_1)))
```

```python
#Find the sum of all elements in an array.
print(np.sum(array_1))
```

```python
#Get elements greater than 3 from [1,2,3,4,5]
array = np.array([1,2,3,4,5])
msk = array > 3

#printing the array..
print("Arrays:",array[msk])
```

```python
#Create a 3×3 identity matrix.
np.eye(3)
```

```python
#Generate 5 random numbers between 0 and 1
random_num = np.random.rand(5)

#printing the num..
print("Random Nums:",random_num)
```

```python
#Convert a 2D array into a 1D array
arr2D = np.array([[1,2,3],
                  [4,5,6]])

#Convert the array in 1D array..
arr1D = arr2D.flatten()

#printing the array..
print("Printing the Flatten array:",arr1D)
```
