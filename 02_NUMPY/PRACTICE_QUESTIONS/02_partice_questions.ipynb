```python
import numpy as np

arr = np.array([1,2,3,4,5])

#average array
avr = np.mean(arr)
print("avr:",avr)

#min of array
min_val = np.min(arr)
print("min_val:",min_val)

#max of array
max_val = np.max(arr)
print("max_val:",max_val)
```

```python
#create a 3x3 numpy array with value from 10 to 18.
array = np.arange(10,19)

#reshape the array...
reshape_arr = array.reshape((3,3))

#printing the array...
print("ARRAYS:",reshape_arr)
```

```python
#FIND MEAN, MEDIAN AND STD OF AN ARRAY...
arr = np.arange(1,13)

#mean of array..
mean_arr = np.mean(arr)
print("Mean:",mean_arr)

#median of array
median_arr = np.median(arr)
print("Median:",median_arr)

#std of array
std_arr = np.std(arr)
print("STD:",round(std_arr,2))
```

```python
#replace all even number with zero
arr = np.arange(1,13)

repalced_arr = arr[arr % 2 == 0] = 0
print(arr)
```

```python
#abstract elements greter than 15.
arr = np.array([1,23,4,56,78,8])

msk = arr > 15
print("Element Gretter Than 15:",arr[msk])
```

```python
#How do you create 2D numpy array and compute row wise and column wise sum.
array = np.array([[1,2,3],
                 [4,5,6]])

sum_row = np.sum(array, axis=1)
sum_colm = np.sum(array, axis=0)

print("Sum of Row:",sum_row)
print("Sum of col:",sum_colm)
```

```python
#Any Example of Broadcasting example:
arr1 = np.arange(1,6)
arr2 = np.arange(6,11)

#Broadcasting Array...
broadcast_arr = arr1 + arr2

#Printing the Arrays...
print("Sum of Array:",broadcast_arr)
```
