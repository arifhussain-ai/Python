```python
#Reshapeing and flattening arrays...
arrays = np.array([[1,2,3],[4,5,6]])
print("Shape:",arrays.shape)

#Reshape to 3,2....
reshape_arr = arrays.reshape((3,2))
print("Reshape:",reshape_arr.shape)

#Flattening Array....
flatten = arrays.flatten() #Converted the 2D array to 1D..
print("Flattend_Array:",flatten)
```
