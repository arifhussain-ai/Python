```python
#USING PYHTON LIST....
import numpy as np
import time

size = 1_000_000_0
l1 = list(range(size))
l2 = list(range(size))

start = time.time()
add = [ x+y for x,y in zip(l1,l2)]
end = time.time()
print(add[0:10])
print("Time_Taken:",end - start)
```

```python
#USING NUMPY ARRAY....
import numpy as np
import time

size = 1_000_000_0
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))

start = time.time()
add = l1 + l2
end = time.time()
print(add[0:10])
print("Time_Taken:",end - start)
```

```python
#NumPy arrays use significantly less memory compared to Python lists.
import sys

list_size = list(range(1000))
numpy_size = np.array(list_size)

print("Python_Size:", sys.getsizeof(list_size) * len(list_size), "Bytes")
print("Numpy_Size:", numpy_size.nbytes, "Bytes")
```

```python
import time

#Pyhton list loop....
list1 = [1,2,3]
start = time.time()
list_square = [x**2 for x in list1]
end = time.time()
print(list_square)
print("Time_Taken:",end - start)

# numpy (Vectorization)
arr1 = np.array([1,2,3])
start = time.time()
numpy_sqaure = arr1 ** 2
end = time.time()
print(numpy_sqaure)
print("Time_Taken:",end - start)
```
