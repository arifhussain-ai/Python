```python
#01_Create a NumPy array with values from 10 to 100 and print its shape.
import numpy as np

arr = np.arange(10, 101, 1)

#Print Array and its shape...
print("Arrays:",arr)
print("Shape:",arr.shape)
```

```python
#02_Compare the time taken to multiply two Python lists vs. two NumPy arrays.

import numpy as np
import time

#Multiply the two pyhton list....
l1 = [1,2,3,4]
l2 = [5,6,7,8]

start = time.time()
multi_lst = [x*y for x,y in zip(l1,l2)]
end = time.time()

#Print Time and Result...
print("List_Multiply:",multi_lst)
print("Time_Taken:",end-start)

#Multiply the tw array....
arr1 = np.array(list(l1))
arr2 = np.array(list(l2))

start = time.time()
multi_arr = arr1 * arr2
end = time.time()

#Print Time and Result...
print("Array_Multiply:",multi_arr)
print("Time_Taken:",end-start)
```
