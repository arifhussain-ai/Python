```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent":
"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}
```

```python
#Data
all_books = []
for i in range(1,51):
    url = f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url,headers = headers)

    soup = BeautifulSoup(response.text,"html.parser")
    #Step 5 — Extract Product Names
    products_name = soup.find_all("article", class_ = "product_pod")
    
    for product in products_name:
        title = product.h3.a["title"]
        prices = product.find("p",class_= "price_color").text.split("Â")[1]
        rateings = product.select_one("p.star-rating")["class"][1]
        stock = product.find("p", class_ = "instock availability").text.strip()
        
    #Extract Product Information
        all_books.append({
            "PRODUCT":title,
            "PRICE":prices,
            "RATINGS":rateings,
            "Stock":stock
        })
```

```python
df = pd.DataFrame(all_books)
df
```

```python
df.to_csv("products.csv", index = False)
```
