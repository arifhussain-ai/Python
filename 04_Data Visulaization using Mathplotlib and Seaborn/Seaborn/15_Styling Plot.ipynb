```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style = "darkgrid")
sns.barplot(data = var, x = "day", y ="total_bill")

plt.show()
```

```python
sns.set(style = "whitegrid")
plt.figure(figsize = (10,6))
sns.barplot(data = var, x = "day", y ="total_bill")

sns.despine()
plt.show()
```

```python
sns.set(style = "whitegrid")
sns.set_context("poster", font_scale = 0.6)

sns.barplot(data = var, x = "day", y ="total_bill", palette= "Set2", hue = "day")

plt.show()
```
