```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x = "day", y = "total_bill",hue = "time", data = var, palette="Dark2")

plt.show()
```

```python
sns.violinplot(x = "time", y = "total_bill",hue = "time",order= ["Dinner","Lunch"],inner = "stick",
               data = var, palette="Dark2")

plt.show()
```

```python
sns.violinplot(x = "day", y = "total_bill",hue = "sex",split= True,density_norm ="width", data = var, palette="Dark2")

plt.show()
```

```python
sns.violinplot(x = "total_bill",y = "day",hue = "sex",split= True,density_norm ="width", data = var, palette="Dark2")

plt.show()
```
