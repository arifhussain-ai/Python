```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("penguins").head(90)

var
```

```python
m = {"Male":"*","Female":"o"}
sns.scatterplot(x= "bill_length_mm",y= "bill_depth_mm",data= var,hue= "sex",style= "sex",
               size= "sex",sizes = (100,150),palette= "gist_ncar", alpha = 0.7,markers=m)

plt.title("Scatter Plot", fontsize = 20, fontweight = "bold")
plt.legend(title = "sex",loc = "lower right")
plt.show()
```
