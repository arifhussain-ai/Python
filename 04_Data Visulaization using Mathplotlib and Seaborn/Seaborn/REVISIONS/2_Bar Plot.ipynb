```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.barplot(data=tips, x="day", y="total_bill",hue = "day",palette= "Accent")

plt.title("Average Total Bill per Day",fontweight = "bold",fontsize = 20)
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.barplot(x= "smoker", y = "total_bill", palette= "coolwarm",hue = "smoker", data = tips)

plt.title("Average Bill: Smokers vs Non-Smokers",fontweight = "bold",fontsize = 16)
plt.show()
```

```python
'''Interview Follow-up:

How to remove confidence interval?'''

sns.set(style = "darkgrid")
sns.barplot(x= "smoker", y = "total_bill", palette= "coolwarm",hue = "smoker",errorbar= None ,data = tips)

plt.title("Remove Confidence Interval",fontweight = "bold",fontsize = 16)
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.barplot(x = "day", y = "total_bill", hue = "sex", data = tips)

plt.title("Average Bill By Day And Genders", fontsize = 18, fontweight = "bold")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

sns.set(style= "darkgrid")
sns.barplot(x = "species", y = "body_mass_g",hue = "species",palette= "Paired" , data = penguins)

plt.title("Average Body Mass By Species", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

sns.set(style = "darkgrid")
sns.barplot(x = "class", y = "survived",hue = "class",palette="cubehelix" ,data = titanic)

plt.title("Survival Rate by Passenger Class", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
flights = sns.load_dataset("flights")

sns.set(style = "darkgrid")
sns.barplot(x = "month", y = "passengers", hue= "month",palette= "Set1", data = flights)

plt.title("Average Monthly Passengers", fontsize = 20, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

tips["tips_percentage"] = (tips["tip"] / tips["total_bill"]) * 100
sns.set(style = "darkgrid")

sns.barplot(x = "day", y = "tips_percentage", palette= "coolwarm",hue = "day", data = tips)

plt.title("Average Tip Percentage by Day",fontweight = "bold",fontsize = 16)
plt.show()
```
