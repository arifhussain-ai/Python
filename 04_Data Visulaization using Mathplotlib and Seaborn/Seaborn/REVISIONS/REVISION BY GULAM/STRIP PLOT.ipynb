```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")

plt.figure(figsize = (8,6))
sns.set(style = "darkgrid")

sns.stripplot(data = tips, x = "day", y = "total_bill", jitter = True, size = 6, 
              alpha = 0.8, palette= "Set2",hue = "sex",edgecolor= "black", linewidth= 1)

plt.title("Customer Spending Pattern by Day", fontsize = 20, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

plt.figure(figsize = (8,6))
sns.set(style = "darkgrid")

sns.stripplot(data = tips, x = "sex", y = "tip", hue = "smoker", alpha = 0.9, dodge= True,
              jitter = True)

plt.legend(title = "Smoker")
plt.title("Tipping Behavior by Gender & Smoking", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

plt.figure(figsize = (8,6))
sns.set(style = "darkgrid")

sns.stripplot(data = penguins, x = "species", y = "body_mass_g", jitter= True, alpha = 0.9,
             hue = "species",size = 5, linewidth= 1, palette= "Accent")

plt.title("Penguin Body Mass by Species", fontsize = 21, fontweight = "bold")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")

plt.figure(figsize = (8,6))
sns.set(style = "darkgrid")

sns.stripplot(data = titanic, x = "survived", y = "age", hue = "sex", dodge= True,
             jitter = 0.24, alpha = 0.8, palette= "muted",size = 5, linewidth= 1)

plt.title("Survival Age Distribution", fontsize = 21, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")

title = sns.catplot(data = tips, x= "day", y= "total_bill", col = "time",kind="strip", hue = "sex",
            jitter  = 0.25,height= 5, aspect= 0.9)

title.fig.suptitle("Separate charts for lunch & dinner", fontsize = 21, fontweight = "bold", y = 1.05)
plt.show()
```
