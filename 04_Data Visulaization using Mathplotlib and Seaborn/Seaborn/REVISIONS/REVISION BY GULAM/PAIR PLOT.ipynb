```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = sns.load_dataset("iris")
sns.set(style = "darkgrid")

sns.pairplot(iris, hue = "species", diag_kind="kde", palette = "Set2", markers= ["o",">","d"],
           x_vars = ["sepal_width", "sepal_length"], y_vars = ["petal_width","petal_length"],
             corner= True, plot_kws= {"alpha": 0.8, "s":60})

plt.suptitle("Iris Feature Relationshop Anlaysis", fontsize = 21, fontweight = "bold", y = 1.03)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
penguins_clean = penguins.dropna()

sns.pairplot(penguins_clean, hue = "species", diag_kind= "hist",height= 4, palette= "deep")
plt.suptitle("Penguin Morphological Analysis", fontsize = 35, fontweight = "bold", y = 1.03)

plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.pairplot(tips, vars = ["total_bill","tip","size"], hue = "smoker", 
             plot_kws= {"alpha":0.8, "edgecolor":"black","s":50})

plt.suptitle("Customer Spending Behavior Analysis", fontsize = 20, fontweight = "bold", y = 1.03)
plt.show()
```

```python
iris = sns.load_dataset("iris")

sns.pairplot(iris, vars = ["sepal_width", "sepal_length","petal_length"], diag_kind= "kde", hue = "species",
            corner= True, palette= "pastel")

plt.suptitle("Selected Pairplot Feature Analysis", fontsize = 20, fontweight = "bold", y = 1.03)
plt.show()
```

```python
iris = sns.load_dataset("iris")

sns.set(style = "darkgrid")

sns.pairplot(iris, diag_kind= "kde",diag_kws= {"fill":True} ,
             plot_kws= {"alpha": 0.8 ,"s": 50}, hue = "species", height= 4, palette= "Accent")

plt.suptitle("Professional EDA dashboard", fontsize = 39, fontweight = "bold", y = 1.02)
plt.show()
```
