```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
sns.set(style = "whitegrid")

sns.histplot(x = "total_bill", bins = [0,10,20,30,40,50,60], hue = "sex", kde = True, element="step",
             palette= "Set2", alpha = 0.7, data = tips)

plt.title("Revenue Distribution by Gender", fontsize = 16, fontweight = "bold")
plt.xlabel("Total Bill", fontsize = 11)
plt.ylabel("Frequency", fontsize = 11)

plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.set(style = "darkgrid")

sns.histplot(x = "total_bill",bins = [0,10,20,30,40,50,60] ,stat="density",kde = True, 
            color = "g", data = tips)

plt.title("Customer Spending Density Distributions", fontsize = 16, fontweight = "bold")
plt.xlabel("Total Bill", fontsize = 11)
plt.ylabel("Density", fontsize = 11)

plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.set(style = "darkgrid")

sns.displot(x = "total_bill", bins =[0,10,20,30,40,50,60], col = "time", kde = True,palette= "Set1",
           hue= "time",height= 8, aspect= 0.6, data = tips)

plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize = (9,5))

sns.histplot(x = "total_bill", bins= [0,10,20,30,40,50,60],cumulative=True, stat="density"
            ,color = "g", data = tips)

plt.title("Cumulative Customer Spending Distributions", fontsize = 19, fontweight = "bold")
plt.xlabel("Total Bill", fontsize = 11)
plt.ylabel("Density", fontsize = 11)

plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.set(style = "darkgrid")

sns.displot(x = "flipper_length_mm",row  = "species", kde = True, hue = "species",height= 4,aspect = 1,
           bins = 10, palette= "Set2",data= penguins)

plt.show()
```
