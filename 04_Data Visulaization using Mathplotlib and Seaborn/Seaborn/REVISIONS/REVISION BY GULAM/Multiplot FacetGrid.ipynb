```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")

fg = sns.FacetGrid(data = tips,row = "sex" ,col = "time", hue = "sex", palette= "muted",
                   height= 4, aspect= 1.2)

fg.map(sns.histplot, "total_bill", kde = True)

fg.add_legend()
fg.set_titles(col_template = "{col_name} Time",
             row_template = "{row_name}")

fg.fig.suptitle("Restaurant Revenue Pattern by Time & Gender", fontsize = 16, fontweight = "bold", y = 1.05)
plt.show()
```

```python
tips = sns.load_dataset("tips")

fg = sns.FacetGrid(data = tips,row = "sex" ,col = "time", hue = "sex", palette= "muted",
                   height= 4, aspect= 1.2)

fg.map(sns.histplot, "total_bill", kde = True)

fg.add_legend()


fg.fig.suptitle("Restaurant Revenue Pattern by Time & Gender", fontsize = 16, fontweight = "bold", y = 1.05)
plt.show()
```

```python
tips = sns.load_dataset("tips")

fg = sns.FacetGrid(data = tips, col = "day", hue = "smoker", palette= "Set2", height= 4 , aspect= 1)

fg.map(sns.scatterplot, "total_bill", "tip", alpha = 0.8)
fg.add_legend()

fg.fig.suptitle("Customer Tip Behavior Across Days & Smoking", fontsize = 21, fontweight = "bold", y = 1.05)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

fg = sns.FacetGrid(data = penguins, hue= "species", col = "island",height=4 , aspect = 1)

fg.map(sns.kdeplot,"body_mass_g",fill  = True)
fg.add_legend()

fg.fig.suptitle("Penguins Body Mass Across Island & Species", fontsize = 18, fontweight = "bold", y = 1.08)

plt.show()
```

```python
titanic = sns.load_dataset("titanic")

fg = sns.FacetGrid(data = titanic, col = "class",row = "survived", hue = "sex", palette= "muted",
                  height= 4 , aspect= 1)

fg.map(sns.histplot,"age", kde = True)
fg.set_axis_labels()
fg.add_legend()

fg.fig.suptitle("Titanic Age Survival Dashboard", fontsize = 25, fontweight = "bold", y = 1.08)

plt.show()
```
