```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize = (9,6))

sns.set(style = "darkgrid")
sns.violinplot(x = "day", y = "total_bill", hue= "sex", split= True, inner="quartile",edgecolor = "black",
               palette= "Set2", density_norm= "width", cut = 0, data = tips)

plt.title("Revenue Distribution By Day & Gender",fontsize = 20, fontweight = "bold")
plt.xlabel("Day",fontsize = 11)
plt.ylabel("Total Bill",fontsize = 11)
plt.show()
```

```python
plt.figure(figsize = (9,6))
sns.violinplot(data = tips, x = "smoker",  y = "tip", palette= "coolwarm", inner= "point"
               ,edgecolor = "black",hue = "smoker", linewidth = 2,bw_method=0.3)

plt.title("Tipping Behavior Distribution", fontsize = 21, fontweight = "bold")
plt.xlabel("Smoker",fontsize = 11)
plt.ylabel("Tip",fontsize = 11)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
plt.figure(figsize = (8,6))

sns.set(style = "darkgrid")
sns.violinplot(data = penguins, x = "species" , y = "body_mass_g", hue = "species", palette= "Accent",
             density_norm="count", inner= "box", edgecolor = "black", linewidth = 2)

plt.title("Penguin Body Mass By Species", fontsize = 21, fontweight = "bold")
plt.xlabel("Species",fontsize = 11)
plt.ylabel("Body Mass",fontsize = 11)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize = (8,6))

sns.set(style = "darkgrid")
sns.violinplot(data = titanic, x = "survived", y = "age", hue = "sex", cut = 0, palette= "pastel",
               split= True,inner= "quartile")

plt.title("Survival Age Distribution", fontsize = 21, fontweight = "bold")
plt.xlabel("Survived",fontsize = 11)
plt.ylabel("Age",fontsize = 11)
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
v = sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex",col = "time",kind="violin",split = True,
           inner = "quartile",height= 5, aspect= 0.6)

v.fig.suptitle("Analyze Revenue Distribution By Day",fontsize = 20 ,fontweight = "bold",y = 1.05,
              color = "brown")
plt.show()
```
