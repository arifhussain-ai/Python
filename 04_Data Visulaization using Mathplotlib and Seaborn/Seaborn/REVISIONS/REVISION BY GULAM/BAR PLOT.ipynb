```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips")

plt.figure(figsize=(10,6))
sns.set(style = "darkgrid")
sns.barplot(x = "day", y = "total_bill", hue = "smoker", 
            errorbar=('ci', 95), palette= "Set2", data = tips)

plt.title("Average Revenue Of Smoker & Non-Smoker", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")
order = tips.groupby("day",observed= True)["total_bill"].mean().sort_values().index

sns.barplot(x = "day", y = "total_bill", order = order, hue = "day" ,
            palette= "coolwarm", errorbar=None, data = tips)

plt.xticks(rotation = 45)
plt.title("Average Sorted Revenue Analysis", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
plt.figure(figsize = (10,6))

sns.barplot(y = "species", x = "body_mass_g", orient = "h", errorbar = None, hue = "species",
            palette= "Accent" ,data = penguins)

plt.title("Average Body Mass By Penguins Species", fontsize = 21, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.barplot(x = "time", y = "tip", hue = "time",estimator= sum, errorbar= None, 
            palette= "cubehelix", data =tips)

plt.title("Total Tip Collected (LUNCH vs DINNER)", fontsize = 16, fontweight = "bold")
plt.xlabel("Time", fontsize = 12)
plt.ylabel("Total Tip", fontsize = 12)
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.catplot(x = "day", y = "total_bill",kind="bar", hue = "day",errorbar=('ci', 95),height=5,
            aspect=1,col = "sex",palette= "Set2", data = tips)

plt.show()
```
