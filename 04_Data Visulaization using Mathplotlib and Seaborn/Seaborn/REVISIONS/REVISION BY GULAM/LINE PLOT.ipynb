```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
fligths = sns.load_dataset("flights")

sns.lineplot(x = "month", y = "passengers", hue = "year", style = "year",dashes = False,
             marker = "o",palette = "Set2",data = fligths)

plt.title("Airline Growth Analysis", fontsize = 18, fontweight = "bold")
plt.legend(title ="Years", loc = "upper left", bbox_to_anchor = [1,1])
plt.xticks(rotation = 45)

plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.lineplot(data=tips,
             x="day",
             y="total_bill",
             hue="sex",
             estimator="mean",
             errorbar=('ci', 95),
             marker="o")

plt.title("Average Daily Bill with Confidence Interval", fontsize = 13, fontweight = "bold")
plt.show()
```

```python
fligths["rolling_avg"] = fligths["passengers"].rolling(12).mean()

sns.set(style = "darkgrid")
sns.lineplot(x = "year",y = "rolling_avg",linewidth = 2, alpha = 0.7,color = "purple" ,data = fligths)

plt.title("12-Month Rolling Average Passenger Growth", fontsize = 13, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.relplot(x = "size",y = "total_bill", col= "sex", kind="line",estimator = "mean", marker = "o",
           data = tips,facet_kws ={"sharey":True})

plt.show()
```

```python
sns.lineplot(x = "year", y = "passengers", marker = "o", data = fligths)

plt.axvline(x = 1960, linestyle = "--", linewidth = 2, color = "red")

plt.title("Highlight Peak Growth Year", fontsize = 18 , fontweight = "bold")
plt.show()
```
