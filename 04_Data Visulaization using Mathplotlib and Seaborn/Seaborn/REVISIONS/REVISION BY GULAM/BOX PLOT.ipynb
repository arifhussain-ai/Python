```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize = (9,6))

sns.set(style = "darkgrid")
sns.boxplot(data = tips, x = "day", y = "total_bill", hue = "sex", dodge= True,
            palette= "pastel", linewidth = 1.5, order=["Fri","Sat","Thur","Sun"],
           fliersize= 7, whis = 1.5)

plt.title("Revenue Distribution & Outlier Detection by Day", fontsize = 20, fontweight = "bold")
plt.xlabel("Day", fontsize = 11)
plt.ylabel("Total Bill", fontsize = 11)
plt.legend(title="Gender")
plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize = (9,6))

sns.set(style = "darkgrid")
sns.boxplot(data = tips, x = "smoker", y = "tip", 
            medianprops = {"marker":"o","color":"red", "linewidth":2},hue = "smoker",
           boxprops = {"edgecolor":"black", "linewidth":1.5}, 
            showfliers = True, whis = 1.5,palette= "Set2")

plt.title("Tip Fairness Analysis Smoker & Non-Smoker", fontsize = 20, fontweight = "bold")
plt.xlabel("Smoker", fontsize = 12)
plt.ylabel("Tip", fontsize = 12)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
plt.figure(figsize = (9,6))

sns.set(style = "darkgrid")
sns.boxplot(data = penguins, x = "species", y = "body_mass_g" , linewidth= 2 , hue = "species",
           whis = 1.5,width= 0.5,showfliers=False)

plt.title("Penguin Weight Stability Analysis", fontsize = 20, fontweight = "bold")
plt.xlabel("Species", fontsize = 12)
plt.ylabel("Body Mass", fontsize = 11)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize = (9,6))

sns.set(style = "darkgrid")
sns.boxplot(data = titanic, x = "survived", y = "age", hue = "sex",
           medianprops = {"color":"brown", "linewidth":2,"marker":"o"},
           dodge = True, linewidth = 2, palette= "Accent")

plt.title("Titanic Survival Age Risk", fontsize = 25, fontweight = "bold")
plt.xlabel("Survived", fontsize = 12)
plt.ylabel("Age", fontsize = 12)
plt.legend(title="Gender")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
v = sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex",kind = "box",
           col = "time", palette= "Set2", height= 5, aspect= 0.8)

v.fig.suptitle("Dashboard Ready Visualization", fontsize = 28, fontweight = "bold", y = 1.06)
plt.show()
```
