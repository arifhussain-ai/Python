```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
plt.figure(figsize = (8,6))

sns.countplot(data = tips, x = "day", hue = "sex", palette= "Accent", dodge= True,
             order= ["Thur","Fri","Sat","Sun"])

plt.title("Customer Count Day By Gender", fontsize = 24, fontweight = "bold")
plt.xlabel("Day", fontsize = 11)
plt.ylabel("Count", fontsize = 11)
plt.legend(title = "Gender")
plt.xticks(rotation=45)

plt.show()
```

```python
tips = sns.load_dataset("tips")
plt.figure(figsize = (8,6))

sns.set(style= "darkgrid")
sns.countplot(data = tips, x = "smoker", hue = "sex", palette= "coolwarm", stat= "percent")

plt.title("Distribution Of Smoking Habit", fontsize = 24, fontweight = "bold")
plt.xlabel("Smoker", fontsize = 12)
plt.ylabel("Percentage", fontsize = 12)

plt.show()
```

```python
titanic = sns.load_dataset("titanic")
plt.figure(figsize = (8,6))

sns.set(style = "darkgrid")
sns.countplot(data = titanic, x = "class", hue = "sex", palette= "coolwarm",
              order= ["First","Second","Third"])

plt.title("Titanic Survival Count Analysis", fontsize = 24, fontweight = "bold")
plt.xlabel("Class", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.legend(title = "Gender")

plt.show()
```

```python
penguins = sns.load_dataset("penguins")

plt.figure(figsize = (8,6))

sns.set(style = "darkgrid")
sns.countplot(data = penguins, x = "species", hue = "island", dodge= True, palette= "pastel")

plt.title("Penguin Species Distribution by Island", fontsize = 20, fontweight = "bold")
plt.xlabel("Species", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.legend(title = "Island")

plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.catplot(data = tips, x = "sex", col = "time", hue = "sex" ,kind = "count", 
            palette= "pastel", height= 5, aspect= 1)

plt.show()
```
