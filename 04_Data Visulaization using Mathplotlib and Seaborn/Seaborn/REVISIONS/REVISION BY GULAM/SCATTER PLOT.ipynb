```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips").head(90)
plt.figure(figsize = (10,6))
sns.set(style = "darkgrid")

sns.scatterplot(x = "total_bill", y = "tip", hue = "day", size = "size", style = "time",
                alpha = 0.8, sizes = (50,200), palette= "Set2", data = tips)

plt.legend(loc = "upper left", bbox_to_anchor= [1,1])

plt.title("Customer Spending vs Tip Analysis", fontsize = 21, fontweight = "bold")
plt.xlabel("Total Bill", fontsize = 12)
plt.ylabel("Tip", fontsize = 12)
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
plt.figure(figsize = (10,10))
sns.set(style = "darkgrid")

sns.scatterplot(x = "flipper_length_mm", y = "body_mass_g", hue = "species",palette= "Accent",
               style = "sex", alpha = 1, data  = penguins)

plt.legend(loc = "upper left", bbox_to_anchor = [1,1])
plt.title("Flipper Length And Body Mass & Gender", fontsize = 21, fontweight = "bold")
plt.xlabel("Flipper Length", fontsize = 13)
plt.ylabel("Body Mass", fontsize = 13)
plt.show()
```

```python
tips = sns.load_dataset("tips").head(90)
plt.figure(figsize=(8,5))

sns.regplot(data = tips,x = "total_bill", y = "tip",scatter_kws={"alpha": 0.8, "color":"g"},ci = 95,
            line_kws={"color":"r","linewidth":2})

plt.title("Regression Analysis: Bill vs Tip", fontsize = 21, fontweight = "bold")
plt.xlabel("Tips", fontsize = 13)
plt.ylabel("Total Bill", fontsize = 13)
plt.show()
```

```python
tips= sns.load_dataset("tips")
sns.set(style = "darkgrid")


sns.relplot(x = "total_bill",y = "tip",col = "sex",kind="scatter", 
            hue = "smoker", height = 5, aspect = 1, data = tips)

plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.set(style = "darkgrid")

sns.pairplot(hue = "species",diag_kind= "kde", palette= "Set2", data = iris)

plt.show()
```
