```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
flights = sns.load_dataset("flights")

plt.figure(figsize = (10,6))
pivt = pd.pivot(index= "year", columns= "month", values= "passengers", data = flights)

sns.heatmap(pivt,annot= True,fmt = "d" ,cmap="coolwarm",linewidths=0.5,
            linecolor="white")
plt.title("Airline Passenger Seasonal Heatmap", fontsize = 18, fontweight = "bold")

plt.show()
```

```python
iris= sns.load_dataset("iris")

plt.figure(figsize = (10,6))
corr_matrix = iris.drop(columns= ["species"]).corr()

sns.heatmap(corr_matrix, annot= True, cmap = "coolwarm",linewidth = 0.5,linecolor = "white")

plt.title("Iris Feature Correlation Matrix", fontsize = 21, fontweight = "bold")
plt.show()
```

```python
iris = sns.load_dataset("iris")
corr_matrix = iris.drop(columns= ["species"]).corr()

mask = np.triu(np.ones_like(corr_matrix, dtype= bool))
plt.figure(figsize = (8,6))

sns.heatmap(corr_matrix, mask = mask, annot= True, cmap = "coolwarm",square=True, linewidth = 0.6)

plt.title("Masked Upper Triangle", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
iris = sns.load_dataset("iris")
corr_matrix = iris.drop(columns= ["species"]).corr()

v =sns.clustermap(corr_matrix, annot = True, cmap = "coolwarm", linewidth = 0.7,
                  linecolor = "black",figsize = (7,6))

plt.title("Cluster Heatmap", fontsize = 19, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

pivt_tips = pd.pivot_table(data = tips,index = "day", columns= "time", observed= True,
                           values = "total_bill", aggfunc= "mean")

sns.heatmap(pivt_tips, cmap = "coolwarm", linewidth = 0.7, linecolor = "black", annot= True, fmt = ".2f")

plt.title("Average Revenue By (Day & Time)",fontsize = 13, fontweight = "bold") 
plt.show()
```
