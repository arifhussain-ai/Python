```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Years" : [2023]* 12 + [2023] * 12,
    "Months" : list(range(1,13)) * 2,
    "Regions" : ["North"] * 12 + ["South"] * 12,
    "Sales" : [200,220,250,270,300,320,310,330,350,370,390,400,
              180,210,230,260,280,300,295,310,330,340,360,380],
    "Profit" : [20,25,30,35,40,45,42,48,50,55,60,65,
               15,20,25,30,35,38,36,40,45,48,52,58]
})

plt.figure(figsize = (10,6))
sns.set_style("whitegrid")

sns.lineplot(x = "Months", y = "Sales", hue = "Regions", style = "Regions",
            markers = True ,data = df)

plt.xlabel("Months", fontsize = 12, fontstyle = "italic")
plt.ylabel("Sales", fontsize = 12, fontstyle = "italic")
plt.title("Monthly Sales Trend by Region (2023)", fontsize = 20, fontweight = "bold")

plt.xticks(range(1,13))
plt.show()
```

```python
data = {
    "CustomerID": range(1, 21),
    "Age": [19, 21, 20, 23, 31, 22, 35, 23, 64, 30,
            67, 35, 58, 24, 37, 22, 35, 20, 52, 35],
    "Gender": ["Male", "Male", "Female", "Female", "Male",
               "Female", "Female", "Male", "Male", "Female",
               "Male", "Female", "Female", "Male", "Male",
               "Female", "Female", "Male", "Male", "Female"],
    "Annual Income": [15, 15, 16, 16, 17, 17, 18, 18, 19, 19,
                      20, 20, 21, 21, 22, 22, 23, 23, 24, 24],
    "Spending Score": [39, 81, 6, 77, 40, 76, 6, 94, 3, 72,
                       14, 99, 15, 77, 13, 79, 35, 66, 29, 98]
}
df = pd.DataFrame(data)

plt.figure(figsize = (10,6))
sns.set_style("whitegrid")

sns.scatterplot(x = "Annual Income", y = "Spending Score",size = "Age"
                ,palette= "deep", sizes = (50,300),hue = "Gender", data = df)

plt.xlabel("Annual Income (k$)", fontsize = 12, fontstyle = "italic")
plt.ylabel("Spending Score (1-100)", fontsize = 12, fontstyle = "italic")
plt.title("Customer Spending Analysis", fontsize = 20, fontweight = "bold")

plt.legend(title = "Customer Info",loc = "upper left" ,bbox_to_anchor = (1, 1))
plt.tight_layout()
plt.show()
```

```python
data = {
    "EmployeeID" : range(1,31),
    "Department" : ["HR"] * 10 + ["IT"] * 10 + ["Finance"] * 10,
    "Salary" : [ 
        40000, 42000, 39000, 41000, 45000, 43000, 44000, 46000, 47000, 80000,   # HR (outlier: 80000)
        60000, 62000, 58000, 61000, 59000, 63000, 64000, 65000, 66000, 120000, # IT (outlier: 120000)
        50000, 52000, 51000, 53000, 54000, 55000, 56000, 57000, 58000, 90000  # Finance (outlier: 90000)
    ] 
}

df = pd.DataFrame(data)

plt.figure(figsize = (9,6))
sns.set_style("whitegrid")
custom_palette = ["#4C72B0", "#55A868", "#C44E52"]

sns.boxplot(data = df, y = "Salary", x = "Department", hue = "Department", palette= custom_palette)

plt.ylabel("Salary", fontsize = 12, fontstyle = "italic")
plt.xlabel("Department", fontsize = 12, fontstyle = "italic")
plt.title("Employee Salary Distribution", fontsize = 20, fontweight = "bold")

plt.show()
```

```python
data = {
    "Marketing Spend": [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    "Sales Revenue":   [25, 30, 45, 50, 65, 70, 85, 95, 110, 120]
}

df= pd.DataFrame(data)

plt.figure(figsize= (9,6))
sns.set_style("whitegrid")

sns.regplot(x = "Marketing Spend", y = "Sales Revenue", data = df)
plt.xlabel("Marketing Spend", fontsize = 12, fontstyle = "italic")
plt.ylabel("Sales Revenue", fontsize = 12, fontstyle = "italic")
plt.title("Marketing Campaign Effectiveness", fontsize = 20, fontweight = "bold")

plt.show()

print(df.corr())
```

```python
data = {
    "Category": [
        "Electronics","Electronics","Electronics",
        "Clothing","Clothing","Clothing",
        "Furniture","Furniture","Furniture",
        "Sports","Sports","Sports"
    ],
    "Units Sold": [
        200, 220, 250,
        150, 170, 160,
        180, 190, 210,
        140, 130, 150
    ]
}

df = pd.DataFrame(data)

plt.figure(figsize= (9,6))
sns.set_style("whitegrid")

category_order = df.groupby("Category")["Units Sold"].mean().sort_values(ascending= False).index

sns.barplot(x = "Category", y = "Units Sold",order= category_order, errorbar= ("ci", 95) ,
                hue = "Category",palette= "Set1" ,data = df)

plt.xlabel("Product Category", fontsize = 12, fontstyle = "italic")
plt.ylabel("Average Units Sold", fontsize = 12, fontstyle = "italic")
plt.title("Average Units Sold by Product Category", fontsize = 20, fontweight = "bold")

plt.show()
```

```python
data = {
    "Sales": [200, 220, 250, 270, 300, 320, 350, 370, 390, 410],
    "Profit": [20, 25, 30, 35, 40, 45, 50, 55, 60, 65],
    "Marketing Spend": [50, 60, 65, 70, 80, 85, 90, 100, 110, 120],
    "Customer Count": [100, 120, 130, 150, 170, 180, 200, 220, 240, 260]  
}

df = pd.DataFrame(data)

plt.figure(figsize = (9,6))
sns.set_style("whitegrid")

corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot= True, fmt = ".3f", linewidth = 0.5,cmap = "coolwarm")

plt.title("Correlation Matrix Heatmap", fontsize = 20, fontweight = "bold")
plt.show()
```
