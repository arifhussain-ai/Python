```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset("tips")

avg_bill = tips.groupby("day",observed=True )["total_bill"].mean().reset_index()

sns.set(style = "darkgrid")
sns.lineplot(x = "day", y = "total_bill", data  = avg_bill, markers="o")

plt.title("Average Total Bill per Day", fontsize = 18, fontweight = "bold")
plt.show()
```

```python
flights = sns.load_dataset("flights")

sns.set(style = "darkgrid")
sns.lineplot(x = "month", y = "passengers",hue = "year", data = flights, palette="Set1")

plt.title("Monthly Passenger Trend By Year", fontsize = 18, fontweight = "bold")
plt.legend(title = "Years",loc = "upper left",bbox_to_anchor = [1,1])
plt.show()
```

```python
Fmri = sns.load_dataset("fmri")

sns.lineplot(x = "timepoint", y = "signal" ,hue = "event", palette= "Accent" ,
            markers = True,data = Fmri)

plt.title("Brain Signal Over Time by Event", fontsize = 18, fontweight = "bold")
plt.show()
```

```python
Penguins = sns.load_dataset("penguins")
avg_flip_length = Penguins.groupby("species")["flipper_length_mm"].mean().reset_index()

sns.lineplot(x = "species", y = "flipper_length_mm", color = "g" ,marker= "o",data = avg_flip_length)

plt.title("Average Flipper Length by Species", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.lineplot(x = "size", y = "total_bill", marker = "o",errorbar="ci" ,data = tips)

plt.title("Average Total Bill by Group Size with CI", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
flights["rolling_avg"] = flights["passengers"].rolling(3).mean()

sns.lineplot(data=flights, x=flights.index, y="rolling_avg")

plt.title("3-Month Rolling Average of Passengers", fontsize = 16, fontweight = "bold")
plt.xlabel("index",fontsize = 12)
plt.show()
```

```python
tips = sns.load_dataset("tips")

tips["tip_percentage"] = (tips["tip"] / tips["total_bill"]) * 100

sns.set(style = "darkgrid")
sns.lineplot(x = "day", y = "tip_percentage",color ="g",marker= "o" ,data  = tips)

plt.title("Average Tip Percentage by Day", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.lineplot(x = "day", y = "total_bill",style = "smoker" ,hue = "smoker",markers = ["o",">"]
             ,dashes = False,data = tips)

plt.title("Average Tip Bill Trend: Smoker vs Non-Smoker", fontsize = 15, fontweight = "bold")
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.set(style = "darkgrid")
sns.lineplot(x = "day", y = "tip", hue = "sex", style = "sex",dashes = False
             ,markers= ["o",">"], data =  tips)

plt.title("Average Tip Trend By Gender", fontsize = 15, fontweight = "bold")
plt.show()
```

```python
flights = sns.load_dataset("flights")

sns.set(style = "darkgrid")
sns.lineplot(x = "month",y = "passengers" ,hue = "year",palette= "Accent",
             dashes= False,style = "year", data = flights)

plt.legend(title = "Years",loc = "upper left", bbox_to_anchor = [1,1])
plt.title("Seasonal Passenger Growth Pattern", fontsize = 17, fontweight = "bold")
plt.show()
```
