```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips").head(60)
var
```

```python
sns.pairplot(var,hue = "sex", hue_order = ["Female", "Male"])

plt.show()
```

```python
sns.pairplot(var,vars = ["total_bill","tip"],hue = "sex", hue_order = ["Female", "Male"])

plt.show()
```

```python
sns.pairplot(var,x_vars = ["total_bill","tip"],hue = "sex", hue_order = ["Female", "Male"])

plt.show()
```

```python
sns.pairplot(var,y_vars = ["total_bill","tip"],hue = "sex", hue_order = ["Female", "Male"])

plt.show()
```

```python
sns.pairplot(var,hue = "sex", hue_order = ["Female", "Male"], kind= "reg")#Sactter,kde,reg,hist

plt.show()
```

```python
sns.pairplot(var,hue = "sex", hue_order = ["Female", "Male"],markers= ["*",">"])

plt.show()
```
