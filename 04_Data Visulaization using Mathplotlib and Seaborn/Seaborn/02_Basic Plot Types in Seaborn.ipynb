```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
plt.figure(figsize = (10,6))
sns.set_theme(style = "dark")
sns.lineplot(x = "total_bill", y = "tip", data = tips)
plt.xlabel("Total Bill", fontsize = 13)
plt.ylabel("Tips", fontsize = 13)
plt.title("Line Plots", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_theme(style = "dark")
sns.scatterplot(x = "total_bill", y = "tip", data = tips)
plt.xlabel("Total Bill", fontsize = 13)
plt.ylabel("Tips", fontsize = 13)
plt.title("Scatter Plots", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_theme(style = "dark")
sns.barplot(x = "total_bill", y = "tip", data = tips)
plt.xlabel("Total Bill", fontsize = 13)
plt.ylabel("Tips", fontsize = 13)
plt.title("Bar Plots", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.set_theme(style = "dark")
sns.boxplot(x = "total_bill", y = "tip", data = tips)
plt.xlabel("Total Bill", fontsize = 13)
plt.ylabel("Tips", fontsize = 13)
plt.title("Box Plots", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
pv = flights.pivot(index="month", columns="year", values="passengers")
```

```python
pv
```

```python
sns.heatmap(pv, annot = True ,fmt = "d", cmap = "YlGnBu")
plt.xlabel("Years", fontsize = 12)
plt.ylabel("Months", fontsize = 12)
plt.title("Passengers Details", fontsize = 16, fontweight = "bold")
plt.show()
```

```python
import pandas as pd

df = pd.DataFrame({
    "Ages":[22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "Salarys":[25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000,
45000],
    "Genders":["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})

sns.scatterplot(x = "Ages", y = "Salarys", hue = "Genders", data = df)
plt.title('Salary vs Age Scatter Plot', fontsize = 16, fontweight = "bold")
plt.show()
```
