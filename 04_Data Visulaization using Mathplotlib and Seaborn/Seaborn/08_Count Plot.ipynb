```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = sns.load_dataset("tips")

var
```

```python
sns.set(style = "darkgrid")
sns.countplot(x ="sex",hue ="smoker", data =var, palette="Accent", alpha = 1)

plt.title("Count Plot", fontsize = 20, fontweight = "bold")
plt.show()
```

```python
sns.countplot(y ="sex",hue ="smoker", data =var)

plt.show()
```
