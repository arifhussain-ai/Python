```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
var = np.linspace(1,10,20).reshape(4,5)

var
```

```python
y = {"fontsize":10,"color":"darkblue"}
v = sns.heatmap(var, vmin = 0, vmax = 10, cmap = "coolwarm", annot= True,annot_kws=y,linecolor = "y", 
                linewidth = 1.5,xticklabels= False, yticklabels= False)

v.set(xlabel = "Numbers", ylabel = "Cube Range")
sns.set(font_scale = 9)
plt.show()
```

```python
arr = np.array([["a0","a1","a2","a3","a4"],
              ["b0","b1","b2","b3","b4"]])

arr
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)

var1
```

```python
plt.figure(figsize = (8,6))
ax = sns.heatmap(var1, vmin = 0, vmax = 10, cmap = "coolwarm", annot= arr,annot_kws={"size":12} ,fmt = "s")

ax.figure.axes[-1].tick_params(labelsize=10) 
plt.show()
```
