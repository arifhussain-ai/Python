```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")

var
```

```python
sns.set(style = "darkgrid")

sns.boxplot(data = var , x = "day", y = "total_bill", hue = "sex",showmeans = True,
            meanprops = {"marker":"o","markeredgecolor":"black"},linewidth= 2,
            palette="pastel",order=["Fri","Sun","Thur","Sat"])

plt.show()
```

```python
sns.set(style = "darkgrid")

sns.boxplot(data = var , y = "day", x = "total_bill", hue = "sex",showmeans = True,
            meanprops = {"marker":"o","markeredgecolor":"black"},linewidth= 2,
            palette="pastel",order=["Fri","Sun","Thur","Sat"])

plt.show()
```

```python
sns.set(style = "darkgrid")

sns.boxplot(data = var,showmeans = True,
            meanprops = {"marker":"o","markeredgecolor":"black"},linewidth= 2,
            palette="pastel",orient = "h")

plt.show()
```

```python
sns.set(style = "darkgrid")

sns.boxplot(x = var["total_bill"])

plt.show()
```

```python
sns.set(style = "darkgrid")

sns.boxplot(y = var["total_bill"])

plt.show()
```
