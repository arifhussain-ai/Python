```python
import matplotlib.pyplot as plt
```

```python
activities = ["Studying","Playing","Watching TV","Sleeping"]
time_spend=[3,2,2,5]
colors=["blue","red","yellow","pink"]

plt.figure(figsize=(6,6))
plt.pie(time_spend, labels=activities, colors=colors, startangle=90, autopct="%1.1f%%")
plt.title("Student Time Table")
plt.axis("equal")
plt.show()
```

```python
days = [1, 2, 3, 4, 5, 6, 7] # Days of the week
studying = [6, 4, 5, 9, 5, 9, 5]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels= ["Studying","Playing","Watching TV","Sleeping"]
colors= ["blue","red","yellow","pink"]
plt.figure(figsize=(6,6))

plt.stackplot(days, studying, playing, watching_tv, sleeping, labels=labels, colors=colors, alpha = 0.4)
plt.legend(loc = "upper left")
plt.title("Student Time Table")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
```

```python
import matplotlib.pyplot as plt

#Datas
days = [1,2,3,4,5,6]
study = [3,3,3,4,2,3]
series = [4,2,5,6,2,5]
gym = [2,3,2,3,1,2]
eating = [1,1,1,1,2,1]
travel = [4,2,1,2,1,1]

#labels and colors
labels= ["Studying", "Web Series", "Workout", "Eating", "Traveling"]
colors= ["blue","red","yellow","pink","green","orange"]

plt.stackplot(days, study, series, gym, eating, travel, labels=labels, colors=colors, alpha= 0.6)
plt.legend(loc= "upper left")
plt.title("My Weekly Schedule")
plt.xlabel("Days")
plt.ylabel("Hours")
plt.grid(True)
plt.show()
```
