```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar Yearly Run Scored")
```

```python
kholi = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sewhag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years,kholi,label="Virat Kholi")
plt.plot(years,sewhag,label="Vireder Sewhag")

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparsion")
plt.legend()
```

```python
plt.plot(years,kholi,"yo--",linewidth = 3,label="Virt Kholi")
plt.plot(years,sewhag,"b^:",linewidth = 2,label="Virender Sewhag")

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Competition")
plt.legend()
```

```python
plt.plot(years, kholi, color="orange",linewidth = 3, linestyle="--", label="Virt Kholi")
plt.plot(years, sewhag, color="green", linestyle=":",linewidth = 2 ,label="Virender Sewhag")
plt.plot(years, runs, color="blue", linestyle="-",linewidth = 2 ,label="Sachin Tendulkar")
plt.grid(True)
plt.style.use('classic')
plt.style.use('ggplot')
plt.legend()
```
