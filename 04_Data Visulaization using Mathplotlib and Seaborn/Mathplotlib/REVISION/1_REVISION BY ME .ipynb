```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("SIMPLE LINE PLOT",fontsize=15)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [i ** 2 for i in x]
y2 = [i ** 3 for i in x]

plt.plot(x, y1, label = "X Squared")
plt.plot(x, y2, label = "X Cubed")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Multiple line")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
subject = ["Math","Science","English"]
marks = [80,60,90]

plt.bar(subject,marks, color = "skyblue")
plt.xlabel("Subjects",color = "red")
plt.ylabel("Marks",color = "red")
plt.title("Showing Scored Marks")
plt.show()
```

```python
x = [5, 7, 8, 6, 9]
y = [10, 14, 12, 11, 15]

plt.scatter(x,y)
plt.xlabel("X Label",color = "red")
plt.ylabel("Y Label", color = "red")
plt.title("Scatter Plot",fontsize = 15)
plt.show()
```

```python
marks = [55, 60, 65, 70, 75, 80, 85, 90, 95]

bins = [50,60,70,80,90,100]
plt.hist(marks, bins= bins,edgecolor = "black")
plt.grid(True)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Histogram of Marks", fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y1 = [i ** 2 for i in x]
y2 = [i * 2 for i in x]

fig, axs = plt.subplots(1,2, figsize = (10,6))

axs[0].plot(x,y1)
axs[0].set_title("X Squared")

axs[1].plot(x,y2)
axs[1].set_title("X * 2")

fig.suptitle("Simple Subplots", fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y,"--o")
plt.grid(True)
plt.xlabel("X Label")
plt.ylabel("Y Label")

plt.title("Styled Line Plot")
plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x,y,"b--o")

plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Plot with Grid")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.figure(figsize=(8,6))

plt.plot(x,y,"b--o")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Custom Figure Szied", fontsize= 15)
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x,y,color = "red", linewidth = 3)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Line Color And Width",fontsize = 15)
plt.show()
```

```python
pruduct = ["A","B","C"]
sales = [85,15,23]

plt.barh(pruduct,sales,color = "skyblue")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.title("Horizontal Bar",fontsize = 15)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 15, 13, 18]

plt.plot(x, y, marker = "o", linestyle = "none")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Marker Without Lines", fontsize = 15)
plt.grid(True)
plt.show()
```

```python
names = ["Jan", "Feb", "Mar", "Apr"]
values = [20, 30, 25, 35]

plt.plot(names, values)
plt.xlabel("NAMES")
plt.ylabel("VALUES")
plt.xticks(rotation = 45)
plt.show()
```

```python
label = ["Rent","Electric bill", "Food bill"]
expense = [2500,500,3000]
explode = [0.1,0,0]
colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"]

plt.pie(expense, labels= label, explode= explode, autopct="%1.1f%%", colors = colors, 
        wedgeprops=  {"edgecolor":"black", "linestyle" : "--", "linewidth" : 2})

plt.title("Expense Distribution", fontsize = 15)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 8, 13, 18]

plt.plot(x,y)

plt.annotate("Highlight Point", xy= (2,8), xytext= (3,13),
             arrowprops=dict(arrowstyle = "->"),color = "red", fontsize = 12)

plt.title("Annotate Point")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x,y)
plt.xticks([1,3,5])
plt.title("Custom Ticks", fontsize = 15)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.step(x,y)
plt.title("Step Plot", fontsize= 15)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x,y)
plt.fill_between(x,y)
plt.title("Area Under Curve", fontsize = 15)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [4, 7, 10, 13]

fig, axs = plt.subplots(1,1, figsize = (9,6))
axs.plot(x,y)
plt.title("Object-Oriented Plot", fontsize = 15)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x = [1,2,3,4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

fig, axs = plt.subplots(2,1, sharex= True)

axs[0].plot(x,y1)
axs[0].set_title("Frist Plot")

axs[1].plot(x,y2)
axs[1].set_title("Second Plot")

fig.suptitle("Share Same X-Axis", fontsize = 15)

plt.tight_layout()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 15, 12, 18]
size = [50,100,200,300]

plt.scatter(x,y)
plt.title("Scatter with Variable Sizes", fontsize = 15)
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 15, 12, 18]

plt.plot(x,y, alpha = 0.5)
plt.title("Transparent Line", fontsize = 15)
plt.show()
```

```python
labels = ["A", "B", "C"]
data1 = [20, 35, 30]
data2 = [25, 15, 20]

plt.bar(labels, data1, label = "Data 1")
plt.bar(labels, data2, bottom = data1, label = "Data2")

plt.title("Stacked Bar Chart")
plt.legend(loc = 'upper left')
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [40,50,60,70]

plt.plot(x,y1,label ='Line 1')
plt.plot(x,y2,label = 'Line 2')
plt.title("Legend Outside Plot")
plt.legend(bbox_to_anchor = (1.05,1), loc = "upper left")
plt.show()
```
