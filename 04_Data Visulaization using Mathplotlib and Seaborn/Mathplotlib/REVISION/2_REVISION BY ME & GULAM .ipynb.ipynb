```python
import matplotlib.pyplot as plt
```

```python
days = list(range(1,31))
traffic = [120,130,125,140,150,160,170,180,175,165,
           155,160,170,185,190,200,210,205,195,185,
           175,180,190,200,210,220,230,225,215,205]

plt.figure(figsize = (10,6))
plt.plot(days,traffic, marker = 'o', linestyle = '--')
plt.xlabel("Days",fontsize = 13)
plt.ylabel("Visitors",fontsize = 13)
plt.title("Daily Website Traffic",fontsize = 15)
plt.grid(True,axis = "y")

max_days = days[traffic.index(max(traffic))]
plt.annotate("Peak Traffic", (max_days,max(traffic)), xytext = (20,210), 
             arrowprops=dict(arrowstyle = "->"), color = "red", fontsize = 13)
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 170, 160, 180, 200]
profit = [30, 45, 50, 48, 55, 65]
customer_satisfaction = [3.8,4.0,4.2,4.4,4.6,4.7]  # percentage
server_downtime = [6,5,4,3,2,1]
employee_count = [50, 52, 55, 58, 60, 62]
marketing_spend = [20, 25, 30, 28, 35, 40]

fig, axs = plt.subplots(3,2, figsize = (10,6))

fig.suptitle("Dashboards", fontsize = 25, fontweight = "bold")

axs[0,0].plot(months, sales)
axs[0,0].set_title("Sales", fontsize = 13)
axs[0,0].grid()

axs[0,1].plot(months, profit)
axs[0,1].set_title("Profits", fontsize = 13)
axs[0,1].grid()

axs[1,0].plot(months, customer_satisfaction)
axs[1,0].set_title("Customer Satisfactions", fontsize = 13)
axs[1,0].grid()

axs[1,1].plot(months, server_downtime)
axs[1,1].set_title("Server Downtimes", fontsize = 13)
axs[1,1].grid()

axs[2,0].plot(months, employee_count)
axs[2,0].set_title("Employee Count", fontsize = 13)
axs[2,0].grid()

axs[2,1].plot(months, marketing_spend)
axs[2,1].set_title("marketing Spend", fontsize = 13)
axs[2,1].grid()


plt.tight_layout()
plt.show()
```

```python
import random

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [200, 220, 250, 270, 300, 330]
employees_experience = [1, 2, 3, 4, 5, 6]
productivity = [3, 6, 10, 15, 21, 28]
performance_scores = [random.randint(50,100) for _ in range(100)]

fig, axs = plt.subplots(2,2, figsize = (10,6))

fig.suptitle("Dashboard", fontsize = 25, fontweight = "bold")

axs[0,0].plot(months, revenue)
axs[0,0].set_title("Revenue", fontsize = 12)
axs[0,0].grid()

axs[0,1].bar(months, employees_experience)
axs[0,1].set_title("Employee Experince", fontsize = 12)
axs[0,1].grid()

sc = axs[1,0].scatter(months, productivity,c=productivity, cmap = "viridis")
fig.colorbar(sc ,ax = axs[1,0] ,label = "Productivity Range")
axs[1,0].set_title("Productivity", fontsize = 12)
#Adding Annotations
for i in range(len(months)):
    axs[1,0].annotate(f"P{i+1}", (months[i],productivity[i]-0.2), ha = "left")
axs[1,0].grid()

axs[1,1].hist(performance_scores, bins = 10)
axs[1,1].set_title("Performance Score", fontsize = 12)
axs[1,1].grid()

plt.tight_layout()
plt.show()
```
