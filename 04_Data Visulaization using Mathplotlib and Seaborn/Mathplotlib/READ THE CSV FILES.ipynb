```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("ODERS.csv")
```

```python
plt.bar(df['order_id'], df['order_amount'])

plt.xlabel('Order Id',color='red')
plt.ylabel('Oder Amount',color='red')
plt.title('Oder Details')
plt.grid(axis = "y")
plt.show()
```

```python
plt.plot(df['order_id'], df['order_amount'],'go--')


plt.xlabel('Order Id',color='red')
plt.ylabel('Oder Amount',color='red')
plt.title('Oder Details')
plt.grid(True)
plt.show()
```

```python
plt.barh(df['order_id'], df['order_amount'])


plt.ylabel('Order Id',color='red')
plt.xlabel('Oder Amount',color='red')
plt.title('Oder Details')
plt.grid(True)
plt.show()
```
