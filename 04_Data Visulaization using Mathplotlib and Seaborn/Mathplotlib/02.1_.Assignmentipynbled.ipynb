```python
import matplotlib.pyplot as plt
import numpy as np

years = [2001,2002,2003,2004,2005]
sachin = [200,100,560,800,300]
kohli = [300,500,250,654,200]
sehwag = [800,200,100,321,250]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, kohli, width=width, label="Kholi")
plt.bar(x + width, sehwag, width=width, label="Sehwag")

plt.ylabel("Runs Scored", color="red")
plt.xlabel("Years", color="red")
plt.xticks(x,years)
plt.legend()
plt.tight_layout()
```

```python
import numpy as np

years = [1994, 1996, 1998, 2000, 2003]

# Runs data
sachin = [1100, 1500, 1800, 1200, 1700]
sehwag = [900, 1400, 1600, 1800, 1500]
kohli  = [500, 800, 1100, 1300, 1500]

x = np.arange(len(years))
width = 0.25

plt.bar(x-width,sachin,width=width,label="Sachin")
plt.bar(x,sehwag,width=width,label="Sehwag")
plt.bar(x+width,kohli,width=width,label="Kohli")

for i in range(len(years)):
    plt.text(x[i] - width, sachin[i] + 20, str(sachin[i]), ha="center")
    plt.text(x[i], sehwag[i] + 20, str(sehwag[i]), ha="center")
    plt.text(x[i] + width, kohli[i] + 20, str(kohli[i]), ha="center")

plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Runs Comparison of Sachin, Sehwag, and Kohli (5 Selected Years)")
plt.xticks(x, years)
plt.tight_layout()
plt.legend()
```

```python
total_runs = [
    sum(sachin),
    sum(sehwag),
    sum(kohli)
]

players = ["Sachin", "Sehwag", "Kohli"]

plt.barh(players, total_runs, color="skyblue")

# Value labels
for i in range(len(players)):
    plt.text(total_runs[i] + 10, i, str(total_runs[i]), va='center')

plt.xlabel("Total Runs in Debut 5 Years")
plt.ylabel("Players")
plt.title("Debut 5-Year Total Runs Comparison")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = [1990, 1992, 1994]
sachin = [500, 700, 1100]
sehwag = [0, 200, 900]

x= np.arange(len(years))
width = 0.3

plt.bar(x - width, sachin, width, label="Sachin")
plt.bar(x + width, sehwag, width, label="Sehwag")

plt.xlabel("Index Values")
plt.ylabel("Runs")
plt.title("Grouped Bar Chart Without plt.xticks()")
plt.legend()
```

```python
years = [1990, 1992, 1994]
runs1 = [500, 700, 1100]
runs2 = [300, 600, 900]

x = np.arange(len(years))
width = 1.2

plt.bar(x,runs1,width,label="Player 1")
plt.bar(x + 0.2,runs2,width,label="Player 2")

plt.xlabel("Index Numbers")
plt.ylabel("Players")
plt.title("Bar Overlap When Width > 1.")
plt.legend()
```
