```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
labels=["Sachin","Sehwag","Kohli","Yuvraj"]
runs=[18000, 8000, 12000, 9500]
colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"]
plt.pie(runs, labels=labels, colors=colors)
plt.title("Career Runs Of Indian Batsman")
plt.show()
```

```python
plt.pie(runs,
        labels=labels,
       colors=colors,
       wedgeprops= {"edgecolor":"#127806","linewidth":2,"linestyle":"--"})
plt.show()
```

```python
explode=[0.1,0,0,0]
plt.pie(runs,labels= labels,colors= colors,explode= explode,shadow= True,startangle= 100,autopct="%1.1f%%")
plt.title("Explode Pie Example")
plt.show()
```

```python
# Too many categories
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust',
'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]
plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
plt.show()
```

```python
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust',
'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]
plt.bar(languages,usage,color="skyblue")
plt.xlabel("Languages")
plt.ylabel("Usages")
plt.title("Programming Languages")
plt.show()
```
