```python
import matplotlib.pyplot as plt
```

```python
Kholi = [20,80,45,96,36,14,21,85,78,10]
Rohit = [34,65,87,9,1,23,45,56,54,76]
Sharma = [43,65,63,76,89,1,0,9,87,54]
years = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]

plt.plot(years,Kholi,"ro--",linewidth =1,label="Virt Kholi")
plt.plot(years,Rohit,"g^:",linewidth =2,label="Rohit Sharma")
plt.plot(years,Sharma,"y-.",linewidth =3,label="Kl Sharma")

plt.style.use('classic')
plt.style.use('ggplot')
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performanced Competitions")
#plt.xkcd()
plt.legend()
```

```python
years = [2015, 2016, 2017, 2018, 2019]
player_a = [300, 450, 500, 600, 750]
player_b = [280, 400, 550, 580, 700]

plt.plot(years, player_a, label="Player A")
plt.plot(years, player_b, label="Player B")
plt.legend()
plt.show()
```

```python
A = [300, 450, 500, 600, 750]
B = [280, 400, 550, 580, 700]
plt.plot(A,"ro--",label="Player A")
plt.plot(B,"g^:",label="Player B")
plt.legend()
```

```python
A = [300, 450, 500, 600, 750]
B = [280, 400, 550, 580, 700]
C = [285, 300, 850, 280, 900]
plt.plot(A,color = "red",linestyle="--",label= "Player A")
plt.plot(B,color = "orange",linestyle=":",label= "Player B")
plt.plot(C,color = "green",linestyle="-.",label= "Player C")
plt.legend()
```

```python
A = [300, 450, 500, 600, 750]
B = [280, 400, 550, 580, 700]
C = [285, 300, 850, 280, 900]
plt.plot(A,color = "red",linestyle="--",linewidth = 2,label= "Main Player")
plt.plot(B,color = "orange",linestyle=":",label= "Side Player")
plt.plot(C,color = "green",linestyle="-.",label= "Part Time Player")
plt.grid(True)
plt.tight_layout()
plt.legend()
```

```python
with plt.xkcd():
    A = [300, 450, 500, 600, 750]
    B = [280, 400, 550, 580, 700]
    years=[2001,2002,2003,2004,2005]
    plt.plot(A,years,color = "red",linestyle="--",linewidth = 2,label= "Main Player")
    plt.plot(B,years,color = "orange",linestyle=":",label= "Side Player")
    plt.ylabel("Years")
    plt.xlabel("Runs Scored")
    plt.title("Performance Comparsion")
    plt.legend()
```
