```python
n = int(input("ENTER A NUMBER: "))
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10

print("REVERSED NUMBER:",rev)
```

```python
n = int(input("ENTER A NUMBER: "))

if n > 1:
    for i in range(2, n):
        if n % i == 0:
            print("NOT PRIME NUMBER")
            break
    else:
        print("PRIME NUMBER")
else:
    print("NOT PRIME NUMBER")
```

```python
n = int(input("ENTER A NUMBER: "))
even = 0
odd = 0

for i in range(1, n+1):
    if i % 2 == 0:
       even += 1
    else:
        odd += 1

print("EVEN:",even)
print("ODD:",odd)        
```

```python
n = int(input("ENTER A NUMBER: "))
sum_div = 0

for i in range(1, n):
    if n % i == 0:
        sum_div += i

if sum_div == n:
    print("PERFECT NUMBER")
else:
    print("NOT PERFECT NUMBER")
```

```python
n = int(input("ENTER A NUMBER: "))
max_digit = 0

while n > 0:
    digit = n % 10
    if digit > max_digit:
        max_digit = digit
    n //= 10

print("MAX DIGIT NUMBER:", max_digit)
```
